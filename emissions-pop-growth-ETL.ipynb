    Population growth vs total greenhouse gas emissions

```python
#
# preparing files
#
import pandas as pd
import matplotlib.pyplot as plt

# file downloads
    # population dataset
        # https://data.worldbank.org/indicator/SP.POP.GROW?view=chart
        # default file name: API_EN.ATM.GHGT.KT.CE_DS2_en_csv_v2_4260456.csv
    # greenhouse emissions dataset
        # https://data.worldbank.org/indicator/EN.ATM.GHGT.KT.CE?view=chart
        # default file name: API_SP.POP.GROW_DS2_en_csv_v2_4251293.csv


# file locations, edit to your particular file structure
populationloc = 'population/population.csv'
greenhouseloc = 'greenhouse/greenhouse.csv'

population = pd.read_csv(populationloc, skiprows = 4)
greenhouse = pd.read_csv(greenhouseloc, skiprows = 4)

# picking columns
# limit the data to a period between 2010 and 2015
population = population[['Country Name', 'Country Code', '2010', '2011', '2012', '2013', '2014', '2015']]
greenhouse = greenhouse[['Country Name', 'Country Code', '2010', '2011', '2012', '2013', '2014', '2015']]

```

```python
population.head()
```

```python
population['Average Population Growth Trend'] = population.iloc[:, 2:7].mean(axis=1)
population.head()
```

```python
greenhouse['Average greenhouse emissions trend'] = greenhouse.iloc[:, 2:7].mean(axis=1)
greenhouse['Total greenhouse emissions in this period'] = greenhouse.iloc[:, 2:7].sum(axis=1)
greenhouse.head()
```
